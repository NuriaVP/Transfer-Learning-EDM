```python
csv = pd.DataFrame(columns = ['img_ID','DB_EMD','EMD'])
```

```python
emd_train = os.listdir('Data Augmentation EMD/train/EMD')
no_emd_train = os.listdir('Data Augmentation EMD/train/NO EMD')
```

```python
for i in emd_train:
    nombre = i
    dbgr = i[0:-4]+'_EMD'
    grado = 'EMD'
    csv = csv.append({
        'img_ID':nombre,
        'DB_EMD':dbgr,
        'EMD':grado
    }, ignore_index = True)
```

```python
for i in no_emd_train:
    nombre = i
    dbgr = i[0:-4]+'_NO_EMD'
    grado = 'NO_EMD'
    csv = csv.append({
        'img_ID':nombre,
        'DB_EMD':dbgr,
        'EMD':grado
    }, ignore_index = True)
```

```python
print(csv)
```

```python
x_train, x_val, y_train, y_val = train_test_split(
    csv['img_ID'], 
    csv['EMD'], 
    test_size = 0.2, 
    random_state = 42, 
    stratify = csv['EMD'].values)
```

```python
for i in x_val:
    posicion = 0
    lista = list(csv['img_ID'])
    lista_emd = list(csv['EMD'])
    
    for e in lista:
        if (i == e):
            posicion = lista.index(e)
            break
            
    sol = lista_emd[posicion]
    carpeta = ''
    
    if sol=='EMD':
        carpeta = 'EMD'
    else:
        carpeta = 'NO EMD'
        
    shutil.move('Data Augmentation EMD/train/'+carpeta+'/'+i, 'Data Augmentation EMD/validation/'+carpeta)
```
