```python
import os
import skimage
import tensorflow as tf
import numpy as np
import cv2
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import plot_model
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
```

**DATOS PARA ENTRENAR -> OCT**

```python
path_oct_train = "Datos EMD/OCT/TRAIN"
```

```python
nombres_oct = os.listdir(path_oct_train)
```

```python
print(nombres_oct)
```

Obtener las imágenes como tensor:

https://www.tensorflow.org/api_docs/python/tf/image/convert_image_dtype

```python
train_ds = []

for e in nombres_oct:
    img = skimage.io.imread(path_oct_train+'/'+e)
    tensor_img = tf.image.convert_image_dtype(img,  dtype=tf.int64, saturate=False, name=None)
    train_ds.append(tensor_img)
```

```python
train_ds = tf.convert_to_tensor(train_ds)
```

```python
print(train_ds)
```

```python
train_labels = []

for e in nombres_oct:
    if e in os.listdir("Datos EMD/OCT/NO EMD"): #0 significa no EMD
        train_labels.append(0)
    elif e in os.listdir("Datos EMD/OCT/EMD"): #1 SIGNIFICA EMD
        train_labels.append(1)
```

```python
tensor_train_labels = tf.convert_to_tensor(train_labels)
```

```python
print(tensor_train_labels)
```

```python
print(len(train_labels)==len(train_ds))
```

**DATOS PARA TEST -> IPhone**

```python
path_iphone_test = "Datos EMD/OCT/TRAIN"
```

```python
nombres_iphone = os.listdir(path_iphone_test)
```

```python
test_ds = []

for e in nombres_iphone:
    img = skimage.io.imread(path_iphone_test+'/'+e)
    tensor_img = tf.image.convert_image_dtype(img,  dtype=tf.int64, saturate=False, name=None)
    test_ds.append(tensor_img)
```

```python
test_ds = tf.convert_to_tensor(test_ds)
```

```python
print(test_ds)
```

```python
test_labels = []

for e in nombres_iphone:
    if e in os.listdir("Datos EMD/iPhone/NO EMD"):
        test_labels.append(0)
    elif e in os.listdir("Datos EMD/iPhone/EMD"):
        test_labels.append(1)
```

```python
tensor_test_labels = tf.convert_to_tensor(test_labels)
```

```python
print(tensor_test_labels)
```

```python
print(len(test_labels)==len(test_ds))
```

**REDIMENSIONAR LAS IMÁGENES**

```python
train_ds = tf.image.resize(train_ds, (150, 150))
test_ds = tf.image.resize(test_ds, (150, 150))
```

**CONSTRUIR UNA MATRIZ PARA ALBERGAR LAS LABELS**

```python
train_labels = to_categorical(train_labels, num_classes=2)
test_labels = to_categorical(test_labels, num_classes=2)
```

**CONSTRUCCIÓN DEL MODELO**

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input

## Loading VGG16 model
base_model = VGG16(weights="imagenet", include_top=False, input_shape=train_ds[0].shape)
base_model.trainable = False ## Not trainable weights

## Preprocessing input
train_ds = preprocess_input(train_ds) 
test_ds = preprocess_input(test_ds)
```
